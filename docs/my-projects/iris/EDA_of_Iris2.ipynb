```python
import pandas as pd # import biblioteki pandas alias pd
import matplotlib.pyplot as plt # import modułu pyplot z biblioteki matplotlib alias plt
import seaborn as sns # import biblioteki seaborn alias sns
from IPython.display import Markdown
```

```python
df = pd.read_csv('25__iris.csv', sep=",")
# Funkcja 'read_csv' wczytuje plik csv o podanej naziwe
# Argument 'sep' określa separator używany w pliku csv
```

## **1. OGÓLNE SPOJRZENIE NA DANE**

```python
df.sample(10)
```

## **WNIOSKI:**
#### Na podstawie analizy wstępnej można stwierdzić, że długość płatka (ang. petal length), mieszcząca się w przedziale od ~1,76 do ~3,75, stanowi najbardziej wyróżniający parametr spośród dostępnych cech, pozwalający na rozróżnienie gatunków irysów.

## **2. ANALIZA ŚREDNICH PARAMETRÓW**

```python
df_setosa = df[df['klasa (class)'] == "Iris-setosa"] # Generowanie DataFrame 
df_versicolor = df[df['klasa (class)'] == "Iris-versicolor"]
df_virginica = df[df['klasa (class)'] == "Iris-virginica"] 
```

### - Setosa

```python
setosa = df_setosa.plot.hist(alpha=0.5, bins=40, label=df_setosa.columns, range=(0, 9), ylim=(0, 40))
setosa.legend(title='Iris-Setosa')
```

### - Versicolor

```python
versicolor  = df_versicolor.plot.hist(alpha=0.5, bins=40, label=df_versicolor.columns, range=(0, 9), ylim=(0, 40))
versicolor.legend(title='Iris-Versicolor')
```

### - Virginica

```python
virginica  = df_virginica.plot.hist(alpha=0.5, bins=40, label=df_virginica.columns, range=(0, 9), ylim=(0, 40))
virginica.legend(title='Iris-Virginica')
```

## **WNIOSKI Z ANALIZY PARAMETRÓW:**

### Długość kielicha (sepal length):
##### - Iris virginica ma zazwyczaj najdłuższe kielichy – ich długość waha się od 5 do 7,8 cm, a najczęściej spotykany rozmiar to 6–6,5 cm.
##### - Iris versicolor mieści się w zakresie 5–7 cm, z najczęstszą wartością około 5,25 cm i dominującym przedziałem 5,5–6,25 cm.
##### - Iris setosa ma najkrótsze kielichy – od 4,4 do 5,8 cm, przy czym najwięcej wartości przypada na przedział 4,8–5 cm.

### Szerokość kielicha (sepal width):
##### - Iris setosa wyróżnia się największą szerokością kielicha – od 2,5 do 4 cm, z przewagą wartości w zakresie 3–3,75 cm.
##### - Dla Iris versicolor i Iris virginica szerokość ta jest bardziej zbliżona – najczęściej mieści się między 2,8 a 3,2 cm.

### Długość płatków (petal length): 
##### - Iris setosa ma bardzo krótkie płatki – od 1 do 1,9 cm, z dominującą długością około 1,4 cm. 
##### - Iris versicolor ma wyraźnie dłuższe płatki – od 3 do 5 cm, przy czym najczęściej pojawia się długość 4,5 cm.
##### - Iris virginica charakteryzuje się najdłuższymi płatkami – od 4,5 do 6,75 cm, z przewagą wartości w zakresie 5–6 cm.

### Szerokość płatków (petal width):
##### - Iris setosa ma najwęższe płatki – od 0,1 do 0,6 cm, z dominującą wartością 0,2 cm.
##### - Iris versicolor osiąga szerokości od 1 do 1,8 cm, a najczęściej spotykane są wartości między 1,3 a 1,5 cm.
##### - Iris virginica ma najszersze płatki – od 1,4 do 2,5 cm, z dominującym przedziałem 1,8–2,1 cm.

## **3. ANALIZA KORELACJI DLA POSZCZEGÓLNYCH GATUNKÓW** 
### a. Iris - Setosa

```python
correlation_matrix_setosa = df_setosa[['długość kielicha (sepal length)', 'szerokość kielicha (sepal width)', 'długość płatka (petal length)', 'szerokość płatka (petal width)']].corr()
sns.heatmap(correlation_matrix_setosa, annot=True, cmap='coolwarm', fmt=".1f")
```

### b. Iris - Versicolor

```python
correlation_matrix_versicolor = df_versicolor[['długość kielicha (sepal length)', 'szerokość kielicha (sepal width)', 'długość płatka (petal length)', 'szerokość płatka (petal width)']].corr()
sns.heatmap(correlation_matrix_versicolor, annot=True, cmap='coolwarm', fmt=".1f")
```

### c. Iris - Virginica

```python
correlation_matrix_virginica = df_virginica[['długość kielicha (sepal length)', 'szerokość kielicha (sepal width)', 'długość płatka (petal length)', 'szerokość płatka (petal width)']].corr()
sns.heatmap(correlation_matrix_virginica, annot=True, cmap='coolwarm', fmt=".1f")
```

## **WNIOSKI**:

#### * Długość kielicha i szerokość kielicha w Iris setosa: Długość kielicha w Iris setosa wykazuje silniejszą korelację ze szerokością kielicha, w porównaniu do Iris versicolor i Iris virginica.
#### * Długość kielicha a rozmiary płatków w Iris setosa: Długość kielicha w Iris setosa jest mniej skorelowana z rozmiarami płatków w porównaniu do Iris versicolor i Iris virginica.
#### * Długość płatka a szerokość płatka w Iris versicolor: W Iris versicolor długość płatka jest najsilniej skorelowana ze szerokością płatka.

#### Dane zawarte w bazie danych iris.csv umożliwiają szczegółową analizę trzech gatunków irysów – Iris setosa, Iris versicolor oraz Iris virginica – w kontekście rozmiarów ich kielichów oraz płatków. Analiza tych danych pozwala na szereg interesujących obserwacji.

#### Pierwszym istotnym wnioskiem jest to, że Iris versicolor wyróżnia się najbardziej regularnym kształtem kwiatu spośród wszystkich analizowanych gatunków. Oznacza to, że w przypadku Iris setosa obserwujemy wyraźną koncentrację w rozmiarach płatków wokół jednej wartości, co może świadczyć o wyższej stabilności cechy w tym gatunku.

#### Podsumowując, analiza danych pokazuje nie tylko różnice w rozmiarach kwiatów, ale także różnorodność w strukturze i regularności poszczególnych gatunków irysów. Iris versicolor prezentuje wyjątkową regularność kształtu, podczas gdy Iris setosa wyróżnia się stabilnością rozmiaru płatków, co może być interesującym punktem wyjścia do dalszych badań nad genetyką i ekologią tych roślin.
